```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0, 2, 0.01)
y = np.sin(4 * np.pi * t)
# Imperative syntax
plt.figure(1)
plt.clf()
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.title('Sine Wave')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

import numpy as np
A=np.arange(0,35,0.5)
B=np.arange(45,94, 0.7)

plt.plot(A,B,"red")
plt.ylabel("Vector B")
plt.xlabel("Vector A")
plt.title("vectores de rango")
plt.grid("true")
plt.axis([-1,35,40,100])

# Shade the area between y1 and line y=0
plt.fill_between(A, B, 0,
facecolor="red", # The fill color
color='blue', # The outline color
alpha=0.3) # Transparency of the fill


plt.show()
```
