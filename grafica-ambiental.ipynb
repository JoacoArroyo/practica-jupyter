```python
import pylab as plt
import numpy as np
plt.style.use('ggplot')
fig = plt.figure(1)
ax = plt.gca()
# make some testing data
x = np.linspace( 0, np.pi, 1000 )
test_f = lambda x: np.sin(x)*3 + np.cos(2*x)
# plot the test data
ax.plot( x, test_f(x) , lw = 2)
# set the axis labels
ax.set_xlabel(r'$x$', fontsize=14, labelpad=10)
ax.set_ylabel(r'$f(x)$', fontsize=14, labelpad=25, rotation=0)
# set axis limits
ax.set_xlim(0,np.pi)
plt.draw()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A=np.arange(10,40,0.3)
B=np.arange(5,0,-0.05)
C=np.arange(4,1,-0.03)
D=np.arange(15,35,0.2)
E=np.arange(10,25,0.15)
F=np.arange(3.0,0.5,-0.025)

fig,axes=plt.subplots(2,2,figsize=(8,6),sharex=True,sharey=True)
## suptitle("comparacion de pendientes", fontsize=15)
plt.ylabel("Altura (km)")
plt.xlabel("Temperatura (K)")
plt.grid("true")
plt.axis([0,50,0,5])
# Top Left Subplot
axes[0,0].plot(A,B,"red")
axes[0,0].plot(E,F,"purple")
axes[0,0].axis([0,50,0,5])
axes[0,0].set_title("pendiente neutra")
# Top Right Subplot
axes[0,1].plot(A,B,"red")
axes[0,1].plot(A,C,"g--")
axes[0,1].axis([0,50,0,5])
axes[0,1].set_title("pendiente inestable")
# Bottom Left Subplot
axes[1,0].plot(A,B,"red")
axes[1,0].plot(D,B,"y--")
axes[1,0].axis([0,50,0,5])
axes[1,0].set_title("pendientes estable")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A=np.arange(10,40,0.3)
B=np.arange(5,0,-0.05)
C=np.arange(4,1,-0.03)
D=np.arange(15,35,0.2)
E=np.arange(10,25,0.15)
F=np.arange(3.0,0.5,-0.025)

fig,axes=plt.subplots(3,figsize=(8,12),sharex=True,sharey=True)
##sup.title("comparacion de pendientes", fontsize=15)

# Top Subplot
axes[0].plot(A,B,"red")
axes[0].plot(E,F,"purple")
axes[0].axis([0,50,0,5])
axes[0].set_ylabel("Altura (km)")
axes[0].set_xlabel("Temperatura (K)")
axes[0].set_title("pendiente neutra")

# Midle Subplot
axes[1].plot(A,B,"red")
axes[1].plot(A,C,"g--")
axes[1].set_ylabel("Altura (km)")
axes[1].set_xlabel("Temperatura (K)")
axes[1].set_title("pendiente inestable")

# Bottom Subplot
axes[2].plot(A,B,"red")
axes[2].plot(D,B,"y--")
axes[2].set_ylabel("Altura (km)")
axes[2].set_xlabel("Temperatura (K)")
axes[2].set_title("pendiente estable")

plt.yticks([])
plt.xticks([])


plt.show()
```
